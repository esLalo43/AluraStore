### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular el ingreso total de cada tienda sumando la columna 'Precio'
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: ${ingreso_tienda_1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda_2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda_3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda_4:,.2f}")
```

# 2. Ventas por categoría

```python

# Agrupar por 'Categoría' y sumar las ventas para cada tienda
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Imprimir los resultados para cada tienda
print("\nVentas por categoría en Tienda 1:")
print(ventas_por_categoria_tienda1)

print("\nVentas por categoría en Tienda 2:")
print(ventas_por_categoria_tienda2)

print("\nVentas por categoría en Tienda 3:")
print(ventas_por_categoria_tienda3)

print("\nVentas por categoría en Tienda 4:")
ventas_por_categoria_tienda4
```

# 3. Calificación promedio de la tienda


```python

calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"\nCalificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identificar productos más y menos vendidos en cada tienda

# Tienda 1
productos_vendidos_tienda1 = tienda['Producto'].value_counts()
top_10_tienda1 = productos_vendidos_tienda1.head(10)
bottom_10_tienda1 = productos_vendidos_tienda1.tail(10)

print("\nProductos más vendidos en Tienda 1:")
print(top_10_tienda1)
print("\nProductos menos vendidos en Tienda 1:")
print(bottom_10_tienda1)

# Visualizar productos más vendidos Tienda 1
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_tienda1.values, y=top_10_tienda1.index, palette="viridis")
plt.title('Top 10 Productos Más Vendidos en Tienda 1')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.show()

# Tienda 2
productos_vendidos_tienda2 = tienda2['Producto'].value_counts()
top_10_tienda2 = productos_vendidos_tienda2.head(10)
bottom_10_tienda2 = productos_vendidos_tienda2.tail(10)

print("\nProductos más vendidos en Tienda 2:")
print(top_10_tienda2)
print("\nProductos menos vendidos en Tienda 2:")
print(bottom_10_tienda2)

# Visualizar productos más vendidos Tienda 2
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_tienda2.values, y=top_10_tienda2.index, palette="viridis")
plt.title('Top 10 Productos Más Vendidos en Tienda 2')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.show()

# Tienda 3
productos_vendidos_tienda3 = tienda3['Producto'].value_counts()
top_10_tienda3 = productos_vendidos_tienda3.head(10)
bottom_10_tienda3 = productos_vendidos_tienda3.tail(10)

print("\nProductos más vendidos en Tienda 3:")
print(top_10_tienda3)
print("\nProductos menos vendidos en Tienda 3:")
print(bottom_10_tienda3)

# Visualizar productos más vendidos Tienda 3
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_tienda3.values, y=top_10_tienda3.index, palette="viridis")
plt.title('Top 10 Productos Más Vendidos en Tienda 3')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.show()

# Tienda 4
productos_vendidos_tienda4 = tienda4['Producto'].value_counts()
top_10_tienda4 = productos_vendidos_tienda4.head(10)
bottom_10_tienda4 = productos_vendidos_tienda4.tail(10)

print("\nProductos más vendidos en Tienda 4:")
print(top_10_tienda4)
print("\nProductos menos vendidos en Tienda 4:")
print(bottom_10_tienda4)

# Visualizar productos más vendidos Tienda 4
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_tienda4.values, y=top_10_tienda4.index, palette="viridis")
plt.title('Top 10 Productos Más Vendidos en Tienda 4')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.show()
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"\nCosto de envío promedio Tienda 1: ${costo_envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: ${costo_envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: ${costo_envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: ${costo_envio_promedio_tienda4:.2f}")
```
